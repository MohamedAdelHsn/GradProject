```python
# Function to remove the HTML tags

def return_RemoveHTMLTags(strr):
    
    # Print string after removing tags
    return re.compile(r'<[^>]+>').sub('', strr)
 

def print_RemoveHTMLTags(strr):
    
    # Print string after removing tags
    print(re.compile(r'<[^>]+>').sub('', strr))
```
